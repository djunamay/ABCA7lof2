```python
import pandas as pd
```

##### iN RNAseq

```python
# load DEGs
data1 = pd.read_csv('../../bulkRNAseq/y622_degs.csv', index_col=0)
data2 = pd.read_csv('../../bulkRNAseq/choline_degs.csv', index_col=0)
data3 = pd.read_csv('../../bulkRNAseq/g2_degs.csv', index_col=0)
```

```python
data1['comparison'] = 'WT vs p.Tyr622*'
data2['comparison'] = 'p.Tyr622* + H2O vs p.Tyr622* + CDP-choline'
data3['comparison'] = 'WT vs p.Glu50fs*3'

# merge dataframes
data = pd.concat([data1, data2, data3])
```

```python
data.drop(columns=['t', 'B', 'AveExpr'], inplace=True)
data.to_csv('../../extended_tables/merged_degs_iN.csv')
```

### iN lipidomics

```python
data1 = pd.read_csv('../../extended_tables/lcms_y622.csv')
data2 = pd.read_csv('../../extended_tables/lcms_choline.csv')
data3 = pd.read_csv('../../extended_tables/lcms_G2.csv')
```
